## 5. Cuantos hombres hay? Y cuantas mujeres?

```python
cantidad_hombres = len(df[df["Gender"] == "Male"])
cantidad_mujeres = len(df[df["Gender"] == "Female"])
print(f"Hay {cantidad_hombres} hombres y {cantidad_mujeres} mujeres")
```

## 6. Cual es la media de las alturas?

```python
media_altura = df["Height"].mean()
print(media_altura)
```

## 7. Cual es la mediana de los pesos?

```python
mediana_pesos = df["Weight"].median()
print(mediana_pesos)
```

## 8. Cual es la moda de los indices?

```python
moda_indices = df["Index"].mode().values[0]
print(moda_indices)
```

## 9. Mostrar el registro de la persona mas pesada

```python
df.iloc[df['Weight'].idxmax()]
```

## 10. Graficar en barras la cuenta de los indices

```python
ax = sns.countplot(x="Index", data=df)
```

## 11. Graficar un scatterplot de altura vs pesos

```python
ax = sns.scatterplot(x="Height", y="Weight", data=df)
```

## 12. Al grafico anterior darle un color a cada clasificacion de body index

```python
plt.figure(figsize=(7, 7))
sns.scatterplot(x="Height", y="Weight", hue="Index", data=df, palette="Set3")
plt.show()
```
