```python
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
```

```python
external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']
```

```python
df = pd.read_csv("data/medicare.csv")
```

```python
df.head(10)
```

```python
df["year_str"] = df["year"].astype("str")
df["year_brand"] = df[['year_str', 'brand_name']].apply(lambda x: ''.join(x), axis=1)
df
```

```python
df_grouped = df.groupby(["brand_name"])["total_spending", "beneficiary_count"].mean().reset_index()
```

```python
df_grouped.head()
```

```python
ax = sns.scatterplot(x="total_spending", y="beneficiary_count", data=df_grouped)
```
